```python
from random import *
from time import *
import matplotlib.pyplot as plt
n_arr=[]
time_arr=[]
for k in range(1, 1000):
  n_arr.append(k)
  start = time()
  l=[randint(1, 100) for i in range(k)]
  for i in range(0, k-1): # Bubble Sort
    for j in range(0, k-i-1):
      if l[j]>l[j+1]:
        a=l[j]
        l[j]=l[j+1]
        l[j+1]=a
  end=time()
  time_arr.append(end-start)
plt.title("Bubble Sort N^2 Time complexity graph")
plt.xlabel('N')
plt.ylabel('Time')
plt.plot(n_arr, time_arr)
plt.show()
```

```python
# Selection Sort

from random import *
from time import *
import matplotlib.pyplot as plt
n_arr=[]
time_arr=[]
for k in range(1, 1000):
  n_arr.append(k)
  start = time()
  l=[randint(1, 100) for i in range(k)]

  for i in range(0, len(l)-1):    # Selection Sort
    indexMin = i
    for j in range(i+1, len(l)):
      if l[j] < l[indexMin] :
        indexMin = j
    l[i], l[indexMin] = l[indexMin], l[i]

  end=time()
  time_arr.append(end-start)

plt.title("Selection Sort N^2 Time complexity graph")
plt.xlabel('N')
plt.ylabel('Time')
plt.plot(n_arr, time_arr, color = 'r')
plt.show()
```

```python
# Insertion Sort

from random import *
from time import *
import matplotlib.pyplot as plt
n_arr=[]
time_arr=[]
for k in range(1, 1000):
  n_arr.append(k)
  start = time()
  l=[randint(1, 100) for i in range(k)]

  for i in range(1, len(l)):    # Insertion Sort
    key = l[i]
    j = i-1
    while j>=0 and l[j] > key:
      l[j+1] = l[j]
      j-=1
    l[j+1] = key

  end=time()
  time_arr.append(end-start)

plt.title("Insertion Sort N^2 Time complexity graph")
plt.xlabel('N')
plt.ylabel('Time')
plt.plot(n_arr, time_arr, color = 'green')
plt.show()
```

```python
# Quick Sort

from random import *
from time import *
import matplotlib.pyplot as plt

def partition(l, low, high):
  pivot = l[low]
  i = low+1
  j = high
  while True:
    while(i<len(l) and l[i] <= pivot):
      i+=1
    while(l[j] > pivot):
      j-=1
    if i < j:
      l[i], l[j] = l[j], l[i]
    if i >= j:
      break
  l[low], l[j] = l[j], l[low]
  return j

def quicksort(l, low, high):
  if low < high:
    partitionIndex = partition(l, low, high)
    quicksort(l ,low, partitionIndex-1)
    quicksort(l ,partitionIndex+1, high)


n_arr=[]
time_arr=[]
for k in range(1, 1000):
  n_arr.append(k)
  start = time()
  l=[randint(1, 100) for i in range(k)]

  quicksort(l, 0, len(l)-1)

  end=time()
  time_arr.append(end-start)

plt.title("Quick Sort N*logN Time complexity graph")
plt.xlabel('N')
plt.ylabel('Time')
plt.plot(n_arr, time_arr, color = 'cyan')
plt.show()
```

```python
# Merge Sort

from random import *
from time import *
import matplotlib.pyplot as plt

def mergeSort(arr):
    if len(arr) > 1:
  
        mid = len(arr)//2
  
        L = arr[:mid]
        R = arr[mid:]
        mergeSort(L)
        mergeSort(R)
  
        i = j = k = 0
        while i < len(L) and j < len(R):
            if L[i] < R[j]:
                arr[k] = L[i]
                i += 1
            else:
                arr[k] = R[j]
                j += 1
            k += 1
  
        while i < len(L):
            arr[k] = L[i]
            i += 1
            k += 1
  
        while j < len(R):
            arr[k] = R[j]
            j += 1
            k += 1

n_arr=[]
time_arr=[]
for k in range(1, 1000):
  n_arr.append(k)
  start = time()
  l=[randint(1, 100) for i in range(k)]

  mergeSort(l)
  end=time()
  time_arr.append(end-start)

plt.title("Merge Sort N*logN Time complexity graph")
plt.xlabel('N')
plt.ylabel('Time')
plt.plot(n_arr, time_arr, color = 'brown')
plt.show()
```

```python
from random import *
from time import *
import matplotlib.pyplot as plt

def countSort(l):
  n = len(l)
  m = max(l)
  count = [0]*(m+1)
  for i in range(n):
    count[l[i]]+=1
  j=0
  for i in range(m+1):
    while count[i]:
      l[j]=i
      j+=1
      count[i]-=1

n_arr=[]
time_arr=[]
for k in range(1, 1001):
  n_arr.append(k)
  start = time()
  l=[randint(1, 100) for i in range(k)]

  countSort(l)
  end=time()
  time_arr.append(end-start)

plt.title("Count Sort N Time complexity graph")
plt.xlabel('N')
plt.ylabel('Time')
plt.plot(n_arr, time_arr, color = 'magenta')
plt.show()
```
